```python
# Importa o Sympy
import sympy as sp

# Importa o Numpy
import numpy as np

# Importa o Matplot lib
import matplotlib.pyplot as plt

# Importa o módulo Toolbox
import Toolbox as T
```

```python
## Matriz

E1,A1,L1 = sp.symbols('E1 A1 L1')
E2,A2,L2 = sp.symbols('E2 A2 L2')
P,Q = sp.symbols('P Q')

F = sp.Matrix([-P,Q])
t1 = E1*A1/L1
t2 = E2*A2/L2

K = sp.Matrix([[t1+t2,-t2],[-t2,t2]])




```

```python
subs1 = {A1:0.01,L1:1,E1:1*10**9}
subs2 = {A2:0.01,L2:2,E2:1*10**9}

subs3 = {P:100,Q:300}

K = K.subs(subs2).subs(subs1)

sp.pprint(K)
sp.pprint(sp.latex(K))

F = F.subs(subs3)
sp.pprint(F)

U = K.inv() * F

sp.pprint(U)
```

```python
#Solução das partes
x,L = sp.symbols('x L')
N1 = (1 - x/L)
N2 = (x/L)

u1 = ((1 - x/L1)*0+(x/L1)*U[0]).subs({L1:1})
sp.pprint(u1)
sp.pprint(sp.latex(u1))


u2 = ((1 - (x-L1)/(L2))*U[0]+((x-L1)/(L2))*U[1]).subs({L1:1,L2:2})
sp.pprint(u2)
sp.pprint(sp.latex(u2))


```

```python
T.prepara_plot(u1,x,0,1,"u1")
T.prepara_plot(u2,x,1,3,"u2")

plt.xlabel('x')
plt.ylabel('u(x)')
plt.grid(True)
plt.legend()
plt.show()
```
